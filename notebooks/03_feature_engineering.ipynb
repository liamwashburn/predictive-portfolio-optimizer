```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("../data/raw/market_data.csv", index_col="Date", parse_dates=True)

data.head()
```

```python
returns = data.pct_change().dropna()

returns.head()
```

```python
features = pd.DataFrame(index=returns.index)

features["SPY_return"] = returns["SPY"]

features["SPY_lag1"] = returns["SPY"].shift(1)
features["SPY_lag2"] = returns["SPY"].shift(2)
features["SPY_lag3"] = returns["SPY"].shift(3)

features.head()
```

### Lag Variables
These Lag varaibles are made to show the effect of the past values of the variable of the current value of the variable. For example, GDP Lag 1 shows the effect of the GDP of the previous year on the current year's GDP.

I am looking at this because I want to see if there is any relationship between the past values of the variables and the current values of the variables. This can help me to understand if there is any lag effect in the data.

```python
features["MA_10"] = data["SPY"].rolling(10).mean()
features["MA_50"] = data["SPY"].rolling(50).mean()

features.head()
```

### Moving Averages
Moving averages smooth price fluctuations and help identify longer-term trends.

Shorter moving averages capture short-term momentum, while longer averages reflect broader market direction.


```python
features["momentum_10"] = data["SPY"] / data["SPY"].shift(10) - 1
features["momentum_30"] = data["SPY"] / data["SPY"].shift(30) - 1

features.head()
```

### Momentum Indicators

Momentum measures the strength of recent price movements.

Assets that have performed well recently often continue to outperform in the short term, a phenomenon known as momentum.

```python
features["volatility_30"] = returns["SPY"].rolling(30).std()

features.head()
```

### Momentum Indicators

Momentum measures the strength of recent price movements.

Assets that have performed well recently often continue to outperform in the short term, a phenomenon known as momentum.

```python
features["target"] = returns["SPY"].shift(-1)

features = features.dropna()

features.head()
```

### Target Variable

The target variable represents the next day's return for the SPY ETF.

Machine learning models will attempt to predict this value using the engineered features.

```python
features.to_csv("../data/processed/features.csv")
```

### Why I Chose These Features

I chose these features because they are commonly used in financial modeling and have been shown to be effective in predicting stock returns. Lag variables can capture the influence of past values on current values, moving averages can identify trends, and momentum indicators can capture the strength of recent price movements. These features provide a comprehensive view of the market dynamics that may influence the target variable.
